# Setting Up Training Environment

```python
# Get data (raw)
import pandas as pd
data = pd.read_pickle('/home/ostap/projects/epic7_win_predictor/saved_files/data_final.pickle')
```

```python
## First Pick - Phase 2 - Hero 2
data_fp_p2_h2 = data[['left_pick_order_1', 'left_pick_order_2', 'left_pick_order_3', 
                      'right_pick_order_1', 'right_pick_order_2', 
                      'left_attack_damage_1', 'left_attack_damage_2', 
                              'left_receive_damage_1', 'left_receive_damage_2', 
                              'right_attack_damage_1', 'right_attack_damage_2',
                              'right_receive_damage_1', 'right_receive_damage_2',
                              'left_attribute_cd_1', 'left_attribute_cd_2',
                              'left_hero_code_1', 'left_hero_code_2',
                              'right_hero_code_1', 'right_hero_code_2',
                              'right_attribute_cd_1', 'right_attribute_cd_2' ,
                              'left_preban_1', 'left_preban_2', 
                              'right_preban_1', 'right_preban_2', 'first_pick', 'is_win']].copy()
```

```python
# delete classes with population less than 1
data_fp_p2_h2 = data_fp_p2_h2[data_fp_p2_h2.groupby('left_hero_code_2')['left_hero_code_2'].transform('size') > 1]
data_fp_p2_h2
```

```python
# OOP API
from pycaret.classification import *
s = ClassificationExperiment()
s.setup(data_fp_p2_h2, target = 'left_hero_code_2',use_gpu = True, session_id = 890)
```

```python
# compare models
best = s.compare_models()
best
```
